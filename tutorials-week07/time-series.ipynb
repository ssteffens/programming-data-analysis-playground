# Time series

Data set where one of the variables consists of time values

https://docs.python.org/3.7/library/datetime.html

```python
# Import datetime module (part of standard Python package)
import datetime as dt

# Import pandas module which contains time series functionalities
import pandas as pd

# Import seaborn 
import seaborn as sns
```

```python
df = pd.read_csv('https://www.met.ie/latest-reports/observations/download/knock/yesterday') 
```

```python
df
```

```python
df['Time']
```

```python
df['datetime'] = pd.to_datetime(df['Time']).dt.time
df.set_index('datetime')
```

```python
sns.lineplot(x='Time', y="Temperature ºC", data=df)
```
