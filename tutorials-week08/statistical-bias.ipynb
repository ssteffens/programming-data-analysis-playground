## **Probability**

```python
x = np.linspace(-3.0, 3.0, 1000)
y = ss.norm.pdf(x, loc = 0.0, scale = 1.0)

xseg = x[np.logical_and(-1.0 < x, x < 1.4)]
yseg = y[np.logical_and(-1.0 < x, x < 1.4)]

plt.plot (x, y, color = 'k', alpha = 0.5)
plt.fill_between(xseg, yseg, color = 'b', alpha = 0.5)
plt.axvline(x = -1.0, color = 'grey', linestyle = ':')
plt.axvline(x = 1.4, color = 'grey', linestyle = ':')

plt.text(0.2, 0.15, 
        r'$P (a \leq X \leq b) = \int_a^b \frac{1}{\sqrt{2\pi\sigma^2} } e^{ -\frac{(x-\mu)^2}{2\sigma^2} } \, dx$',
         horizontalalignment='center', size = 17)

plt.axhline(y = 0.0, color = 'black', linestyle = '-')

plt.title('Probability of $x$ between $a$ and $b$')
```

```python
x = np.linspace(-5.0, 5.0, 10000)
plt.plot(x, ss.norm.pdf(x, loc = 0.0, scale = 1.0), 'k-', lw = 1.0)

xseg = x[np.logical_and(-5.0 <= x, x <= -3.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'y', alpha = 0.5, linewidth = 0)
xseg = x[np.logical_and(-3.0 <= x, x <= -2.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'r', alpha = 0.5, linewidth = 0)
xseg = x[np.logical_and(-2.0 <= x, x <= -1.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'g', alpha = 0.5, linewidth = 0)
xseg = x[np.logical_and(-1.0 <= x, x <= 1.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'b', alpha = 0.5, linewidth = 0, label = '$1 \sigma = 68.27\%$')
xseg = x[np.logical_and(1.0 <= x, x <= 2.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'g', alpha = 0.5, linewidth = 0, label = '$2 \sigma = 95.45\%$')
xseg = x[np.logical_and(2.0 <= x, x <= 3.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'r', alpha = 0.5, linewidth = 0, label = '$3 \sigma = 99.73\%$')
xseg = x[np.logical_and(3.0 <= x, x <= 5.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color = 'y', alpha = 0.5, linewidth = 0)

plt.title('Normal Distribution - Probability Distribution Function')
plt.legend()
```

## **Sampling distribution**

= Distribution of the means of the samples: 

How are the 1000 means of the samples (of 10) distributed, = what is their difference from the overall mean of 0?

```python
# Formats numbers to be printed with 3 decimal places
np.set_printoptions(formatter = {'float': lambda x: '{0:6.3f}'.format(x)})

# Take samples from a normal distribution with mean of 0 and sigma of 1
# Sample size = 10, for example measure 10 people's height
# Simulate 1000 times (nosamps)
sampsize = 10
nosamps = 1000

# Due to the small sample size of 10, mean will probably be different than mean of overall normal distribution 

samp = np.random.standard_normal((nosamps, sampsize))

print(samp)
```

```python
# Calculate the mean of each of the samples
mean = samp.sum(axis=1) / sampsize
print(mean)
```

This shows that the sample is a good reflection of the parameter, so it can be considered unbiased.

```python
# Variance = square of standard deviation
# Typically variance is calculated first, then determine deviation as its square root
# To calculate variance of one sample of 10: 
    # Calculate mean = sum of values / number of values
    # For each value: Subtract mean and then square the result
    # Add all results together
    # Example: Values: 1, 3, 5, 9, 12
    # Mean = (1 + 3 + 5 + 9 + 12/5 = 30/5 = 6
    # Variance = ((1-6)^2 + (3-6)^2 + (5-6)^2 + (9-6)^2 + (12-6)^2)/5
    # = ((-5)^2 + (-3)^2 + (-1)^2 + 3^2 + 6^2)/5
    # = (25 + 9 + 1 + 9 + 36)/5
    # = 80/5
    # = 16
 

vari = (samp - mean[:, np.newaxis])**2
vari = vari.sum(axis = 1) / (sampsize)

# Calculate average of variance
mean_vari = vari.sum() / nosamps
print(mean_vari)

# Basil's correction to achieve unbiased estimator of population: 
print(mean_vari * (sampsize / (sampsize - 1.0)))
```

If variance of population of values is calculated, then take a sample of 10 and calculate variance of the 10 values, this is a biased statistical value. In the example above variance/standard deviation should be 1. 
Basil's correction can be used to achieve unbiased estimate.
